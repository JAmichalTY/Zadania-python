```python
1:  dla pobranych z klawiatury dwóch liczb zwróci ich iloczyn

a = int(input ("Podaj pierwszą liczbę: "))
b = int(input ("Podaj drugą liczę: "))
print ("Ich iloczyn wynosi ", a * b)

2:. sprawdzi, czy podana liczba jest parzysta

a = int(input ("Podaj liczbę: "))
if a % 2 == 0 :
    print ("liczba parzysta")
else :
    print ("liczba nieparzysta")

3: obliczy sumę kwadratów kolejnych liczb naturalnych mniejszych od zadanej wartości.


n = int(input("Podaj wartość n: "))
a = 0
for i in range(1, n):
    a += i**2
print("Suma kwadratów kolejnych liczb naturalnych mniejszych od", n, "jest równa", a)

4: sprawdzi, czy podana liczba jest pierwsza.


a = int(input ("Podaj liczbę: "))
if a < 2:
    print("Ta liczba nie jest pierwsza")
else:
    for i in range (2, int(a ** (1/2)) + 1):
        if a % i == 0:
            print("Ta liczba nie jest pierwsza")
            break
    else:
        print("Jest to liczba pierwsza")

5: znajdzie dzielniki podanej liczby.

a = int(input ("Podaj liczbę całkowitą: "))
print("Dzielniki tej liczby to: ")
if a > 0:
    for i in range(-a, 0):
        if a % i == 0:
            print(i)
    for i in range(1, a + 1):
        if a % i == 0:
            print(i)
elif a < 0:
    for i in range(a, 0):
        if a % i == 0:
            print(i)
    for i in range(1, -a + 1):
        if a % i == 0:
            print(i)
else:
    print("Jest ich niekończenie wiele")

6: zwróci rozkład zadanej liczby na dzielniki pierwsze.

def zadanie_6(n):
    liczby = []
    for i in range(2,n+1):
        while n % i ==0:
            liczby.append(i)
            n //= i
        if n == 1:
            break
    return liczby
n = int(input("Podaj liczbę: "))
print("Dzileniki pierwsze liczby ",n,"to: ",zadanie_6(n))

7: znajdzie najmniejszą wspólna wielokrotność podanych dwóch liczb.

import math

a = int(input("Podaj pierwszą liczbę: "))
b = int(input("Podaj drugą liczbę: "))
c = math.lcm(a, b)
print("Najmniejsz wspólna wielkorotność tych liczb to: ", c)

8:. przedstawi zadaną liczbę w postaci binarnej.

a = int(input("Podaj liczbę całkowitą: "))
b = bin(a)[2:]
print("Liczba", a, "w postaci binarnej to:", b)

9:dla zadanych wartości współczynników trójmianu kwadratowego znajdzie jego miejsca zerowe.


a = int(input("Podaj pierwszą liczbę: "))
b = int(input("Podaj drugą liczbę: "))
c = int(input("Podaj trzecią liczbę : "))
d = b**2 - 4*a*c

if d < 0:
    print("Równanie niema rozwiązania")
elif d == 0:
    x = -b/(2*a)
    print(" Miejsce zerowe to: ",x)
elif d > 0:
    x1 = (-b - (d**(1/2))) / 2 * a
    x2 = (-b + (d**(1/2))) / 2 * a
    print("Miejsca zerowe to: ",round(x1,2) ,"i", round(x2,2))

10.1: dla danego n obliczy n! . Podaj różne rozwiązania tego problemu.

n = int(input("Podaj liczbę naturalną: "))
a = 1
for i in range(1,n + 1):
    a *= i
print ("Silnia z liczby ",n, " to ",a)

10.2: dla danego n obliczy n! . Podaj różne rozwiązania tego problemu.

def silnia(a):
    if a == 0:
        return 1
    else:
        return a * silnia(a-1)

a = int(input("Podaj liczbę naturalną: "))
print("Silnia podanej liczby to: ", silnia(a))

11:oblicza n-tą liczbę Fibonacciego (n duże)

def Fibo(n):
    if n <= 1:
        return n
    else:
        return Fibo(n-1) + Fibo(n-2)

n = int(input("Podaj numer wyrazu ciągu Fibonacciego: "))
print("Wartość tego wyrazu to: ", Fibo(n))

12:. dla dowolnej ilości argumentów zwróci ich skumulowane sumy częściowe

def sumy(*liczby):
    lista = []
    wynik = 0
    for i in liczby:
        wynik += i
        lista.append(wynik)
    return lista

liczby = input("Podaj liczby oddzielone spacjami: ").split()
liczby = [int(i) for i in liczby ]
print(sumy(*liczby))

13: dla dowolnej ilości argumentów liczbowych zwróci ich średnią arytmetyczną

def średnia(*liczby):
    return sum(liczby)/len(liczby)

liczby = input("Podaj liczby oddzielone spacjami: ").split()
liczby = [float(i) for i in liczby ]
print(średnia(*liczby))

14: dla zadanego łancucha znaków zwróci łańcuch ze znakami w odwrotnej kolejności

def znaki(a):
    return a[::-1]

a = input("Podaj znaki: ")
print(znaki(a))

15: dla danego łańcucha znaków zwróci informację ile jest w nim dużych, a ile małych liter.


def znaki(a):
    duze = 0
    male = 0
    for i in a:
        if i.isupper():
            duze += 1
        elif i.islower():
            male += 1

    return (male,duze)

a = input("Podaj znaki: ")
print("Ilość małych znaków to: ",znaki(a)[0] ,"a dużych: ",znaki(a)[1])

16: dla danej tablicy zwróci tablicę z jej unikalnymi elementami

def znaki(a):
    lista = []
    for i in a:
        if i not in lista:
            lista.append(i)
    return lista



a = input("Podaj znaki: ")
print("Znaki które się nie powtarzają to: ",znaki(a))

17: dla zadanego łańcucha znaków zwróci słownik ze statystykami na temat częstości występowania poszczególnych znaków

from collections import Counter

def lista(a):
    ilość = Counter(a)
    return list(a), ilość

a = input("Podaj znaki: ")
print(lista(a)[1])

18:dla zadanego stringa zwróci liste słów, na które można go podzielić.

def słowa(a):
    b = a.split()
    return b
a = input("Podaj zdanie: ")
print("Lista słów: ", słowa(a))

19:dla zadanego string zwróci jego postać w notacji wielbłądziej, np. dla ”ala ma pythona” otrzymamy ’alaMaPythona

def słowa(a):
    b = a.split()
    b = [i.capitalize() for i in b]
    wynik = "".join(b)
    return wynik
a = input("Podaj zdanie: ")
print(słowa(a))

20: dla stringa w postaci wielbłądziej zwróci jego postać ”normalną”, np. dla ’alaMaPythona’ otrzymamy ”ala ma pythona”.


import re
def zamiana(a):
    return re.sub(r'([A-Z])', r' \1', a).lower()
a = input("Podaj zdanie: ")
print(zamiana(a))

21:. zgodnie z metodą babilońską obliczy przybliżenie pierwiastka kwadratowego danej liczby.


def pierw(n):
    a = n/2
    while abs(a * a - n) > 0.0001:
        a = (a + n / a) /2
    return round(a,4)
n = int(input("Podaj liczbę: "))
print(pierw(n))

22: przybliża wartość π (metoda Monte-Carlo).


import random
random.seed()
n = int(input('n = '))
k = 0
for j in range(n):
    x = random.random()
    y = random.random()
    if x * x + y * y <= 1:
        k += 1
p = 4 * k / n
print("pi =", p)

23: dla list a = [1, 2, 3] i b = [4, 5, 6] zwróci wynik c = [(1, 4),(2, 5),(3, 6)].

a = [1, 2, 3]
b = [4, 5, 6]
c = list(zip(a, b))
print(c)

```
